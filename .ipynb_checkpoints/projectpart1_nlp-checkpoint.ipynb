```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizing(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [lemmatizer.lemmatize(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
data['content'] = data['company_profile'] + data['description']+ data['requirements'] + data['benefits']
```

```python
data.head()
```

```python
data['content'] = data['content'].apply(lemmatizing)
```

## Upsampling

```python
def random_oversample_text_data(text_data, labels, minority_label, fraud_count, non_fraud_count):
    minority_indices = [i for i, label in enumerate(labels) if label == minority_label]
    majority_count = non_fraud_count
    minority_count = fraud_count
    
    # Calculate the number of samples to duplicate
    oversample_count = majority_count - minority_count
    
    # Randomly select samples to duplicate
    oversample_indices = random.choices(minority_indices, k=oversample_count)
    
    # Duplicate the selected samples
    for index in oversample_indices:
        # Create a new sample by copying the text data and using the same minority label
        text_data.append(text_data[index])
        labels.append(labels[index])
    
    return text_data, labels
```

```python
# Extract text data and labels from DataFrame
text_data = data['content'].tolist()
labels = data['fraudulent'].tolist()

minority_label = 1  # Fake job postings are labeled as 1

# Perform random oversampling on text data associated with the minority class
text_data_oversampled, labels_oversampled = random_oversample_text_data(text_data, labels, minority_label, fraud_count, non_fraud_count)
```

```python
# Create a DataFrame to display the oversampled data
oversampled_data = {
    'content': text_data_oversampled,
    'fraudulent': labels_oversampled
}
oversampled_data = pd.DataFrame(oversampled_data)
```
